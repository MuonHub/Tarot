```python
%matplotlib inline
import matplotlib.pyplot as plt
#import magpylib as magpy
from numpy import linspace, array, meshgrid, log
import magpylib as magpy
```

```python


if __name__ == '__main__':
#    Dimension of each magnet
    dim = [.5, 1, .5]
    mag = [0, 1, 0]
    pos = [0, 0, 0]
    
    
#    Number of magnets to make.
    num_mags = 16
    theta = .0
    beta = 360 / num_mags
    radius = 5
    show_fields_xy = 0
    show_fields_xz = 0
    show_fields_yz = 0
    
    print('Making mobius using {} magnets spaced {} degrees apart.'.format(num_mags, beta))
    
    mobius = magpy.Collection()
    
    for source in range(0, num_mags, 1):
        
        name = 's'+ str(source)
        
        name = magpy.source.magnet.Box(mag=mag, dim=dim, pos=pos)
        name.rotate(theta/2, [1, 0, 1])
        name.move([radius, 0, 0])
        name.rotate(theta, [0, 1, 0], anchor=[0, 0, 0])
        
        theta = theta + beta
        
        mobius.addSources(name)
        
        

    
#    print(my_trimag.BOM)
    

    fig1 = mobius.displaySystem(suppress=True,direc=True)
    fig1.set_size_inches(8, 8)
    
    if show_fields_xy == True:
    
        xs = linspace(-10,10,20)
        ys = linspace(-10,10,20)
        Bs = array([[mobius.getB([x,y,0]) for x in xs] for y in ys])
        
        fig2, ax = plt.subplots()
        fig2.set_size_inches(8, 8)
        
        X,Y = meshgrid(xs,ys)
        U,V = Bs[:,:, 0], Bs[:,:,1]
        ##ax.streamplot(X, Y, U, V, color=np.log(U**2+V**2), density=1)
        ##ax.streamplot(X, Y, U, V, color=np.log(U**2+V**2), density=1)
        ax.streamplot(X, Y, U, V, color=log(U**2+V**2), density=2)
        ax.streamplot(X, Y, U, V, color=log(U**2+V**2), density=2)
 
    if show_fields_xz == True:
    
        xs = linspace(-10,10,20)
        zs = linspace(-10,10,20)
        Bs = array([[mobius.getB([x,0,z]) for x in xs] for z in zs])
        
        fig2, ax = plt.subplots()
        fig2.set_size_inches(8, 8)
        
        X,Z = meshgrid(xs,zs)
        U,V = Bs[:,:, 0], Bs[:,:,1]
        ##ax.streamplot(X, Y, U, V, color=np.log(U**2+V**2), density=1)
        ##ax.streamplot(X, Y, U, V, color=np.log(U**2+V**2), density=1)
        ax.streamplot(X, Z, U, V, color=log(U**2+V**2), density=2)
        ax.streamplot(X, Z, U, V, color=log(U**2+V**2), density=2)

    if show_fields_yz == True:
    
        ys = linspace(-10,10,20)
        zs = linspace(-10,10,20)
        Bs = array([[mobius.getB([0,y,z]) for y in ys] for z in zs])
        
        fig2, ax = plt.subplots()
        fig2.set_size_inches(8, 8)
        
        Y,Z = meshgrid(ys,zs)
        U,V = Bs[:,:, 0], Bs[:,:,1]
        ##ax.streamplot(X, Y, U, V, color=np.log(U**2+V**2), density=1)
        ##ax.streamplot(X, Y, U, V, color=np.log(U**2+V**2), density=1)
        ax.streamplot(Y, Z, U, V, color=log(U**2+V**2), density=2)
        ax.streamplot(Y, Z, U, V, color=log(U**2+V**2), density=2)

    plt.show()
```
